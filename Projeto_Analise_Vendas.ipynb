# Análise exploratória de KPI's - Faturamento, Quantidade de Produtos Vendidos e Ticket Médio.

1- Importar a base de dados

2 - Calcular o faturamento de cada loja

3- Calcular a quantidade de produtos vendidos em cada loja

4- Calcular o ticket médio de cada loja


1 - Importação e visualização da base de dados

```python
import pandas as pd
tabela_vendas = pd.read_excel('/content/Vendas - Dez (2).xlsx')
display(tabela_vendas)
```

2 - Calcular o faturamento da loja


```python
tabela_faturamento = tabela_vendas[['ID Loja','Valor Final']].groupby('ID Loja').sum()    #slice e agrupamento das colunas por ID das Lojas
tabela_faturamento = tabela_faturamento.sort_values(by='Valor Final', ascending=False)    #Ordenar valores e fazer um Ranking de Faturamento do maior para o menor valor 
display(tabela_faturamento)
```

3- Calcular a quantidade de produtos vendidos em cada loja

```python
tabela_qtd = tabela_vendas[['ID Loja','Quantidade']].groupby('ID Loja').sum() #agrupamento da coluna ID Loja e soma dos valores 
tabela_qtd = tabela_qtd.sort_values( by='Quantidade', ascending=False) #classificação do maior valor para o menor 
display(tabela_qtd)
```

3 - Calcular o ticket médio de cada loja

```python
# ticket médio = faturamento/quantidade
ticket_medio = (tabela_faturamento['Valor Final']/tabela_qtd['Quantidade']) #segmentação da divisão por colunas 
ticket_medio.to_frame() #transforma em tabela 
```
