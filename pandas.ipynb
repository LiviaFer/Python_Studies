##Carregar a base de dados bruta 

```python
vendas_df = pd.read_csv('/Contoso - Vendas - 2017 - Contoso - Vendas - 2017.csv')
produtos_df =pd.read_csv('/Contoso - Cadastro Produtos.csv', sep=';')
clientes_df = pd.read_csv('/Contoso - Clientes.csv', sep=';')
lojas_df = pd.read_csv('/Contoso - Lojas.csv', sep=';' )

display(vendas_df)
display(produtos_df)
display(clientes_df)
display(lojas_df)
```

##Aplicação

##Limpeza de Colunas desnecessárias

```python


vendas_df = vendas_df.drop(['Numero da Venda','Data da Venda','Data do Envio','ID Canal','ID Promocao'], axis=1)
produtos_df = produtos_df[['ID Produto', 'Categoria' ]]
clientes_df = clientes_df[['ID Cliente','E-mail']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja' ]]

display(lojas_df)

```

```python
display(vendas_df)
```

##Agrupar todos os Dataframes

```python

vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')

display(vendas_df)
```

###Renomear o e-mail para ficar claro que é do cliente

```python

vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail do Cliente'})
display(vendas_df)
```

##**Insights**
#Qual cliente comprou mais vezes?

Uso do método value_counts() para contar quantas vezes o mesmo e-mail aparece no Dataframe e plot() para fazer o gráfico.

```python
frequencia_cliente = vendas_df['E-mail do Cliente'].value_counts()
display(frequencia_cliente)
frequencia_cliente[:5].plot(figsize=(15,5)) #largura e altura 
```

##Qual foi a loja que mais vendeu?

Uso  do método groupby para fazer o agrupamento de todas as lojas e sum() para somar todas as unidades vendidas

```python
vendas_loja = vendas_df[['Nome da Loja','Quantidade Vendida']].groupby('Nome da Loja').sum()
display(vendas_loja)
```

Ordenar a quantidade de itens vendidos de forma decrescente usando o método sort_values()

```python
vendas_loja = vendas_loja.sort_values('Quantidade Vendida', ascending= False)
vendas_loja[:5].plot(figsize=(15,5), kind='bar')
```

Qual produto menos vendeu?

Usando os métodos min() para descobrir o menor valor e idxmin() para descobrir o índice do menor valor.

```python
menor_venda = vendas_loja.min()
pior_loja = vendas_loja['Quantidade Vendida'].idxmin()
print(menor_venda, pior_loja)
```

## Qual o percentual de vendas devolvido?
(Total devolvido/total vendas)


```python
qtde_vendidas = vendas_df['Quantidade Vendida'].sum()
print(qtde_vendidas)
```

```python
qtde_devolvida = vendas_df['Quantidade Devolvida'].sum()
print(qtde_devolvida)
```

```python
percentual_devolvido = qtde_devolvida/qtde_vendidas
print('{:.2%}'.format(percentual_devolvido)) #transforma número em porcentagem 
```

##Qual percentual de vendas devolvido somente na loja "Loja Contoso Europe Online	"?


```python

loja_onlineeurope = vendas_df[(vendas_df['ID Loja'] == 306)].sum()
qtde_vendida_europe = loja_onlineeurope['Quantidade Vendida'].sum()
qtde_devolvida_europe = loja_onlineeurope['Quantidade Devolvida'].sum()

percentual_europe = qtde_devolvida_europe/qtde_vendida_europe

print('{:.2%}'.format(percentual_europe))


```

## Quantidade de vendas com zero devoluções na "Loja Contoso Europe Online'


```python
df_loja306semdev = vendas_df[(vendas_df['ID Loja'] == 306) & (vendas_df['Quantidade Devolvida'] == 0 )]
df_loja306semdev.sum()
```
